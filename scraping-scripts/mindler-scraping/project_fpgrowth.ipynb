```python
import json
```

```python
f=open("data.txt","r")
data = ""
temp = f.readlines()
for i in temp:
    data+=str(i).replace('\n','').replace("'",'"')

datastore = json.loads(data)
#print(datastore['data'][0]['Searches'])
data_set=[];
for i in range(len(datastore['data'])):
    data_set.append(datastore['data'][i]['Searches'])
print(data_set)
```

```python
import pyfpgrowth
```

```python
patterns = pyfpgrowth.find_frequent_patterns(data_set, 5)
print(patterns)
```

```python
rules = pyfpgrowth.generate_association_rules(patterns, 0.5)
print(len(rules))
print(rules)
```

```python
A=[]
B=[]
for key, value in rules.items():
    A.append(key)
    B.append(value[0])
    
print(A)
print(B)
```
